```python
import pandas as pd
import yfinance as yf
import plotly.graph_objects as go
```

```python
ticker = yf.Ticker('PETR4.SA')

dados = ticker.history(period='3mo')
dados
```

```python
dados['Media Movel'] = dados['Close'].rolling(window=5).mean().round(2)
dados
```

```python
candlestick = go.Candlestick(x=dados.index, open=dados['Open'], high=dados['High'], low=dados['Low'], close=dados['Close'])

m5 = go.Scatter(x=dados.index, y=dados['Media Movel'], mode='lines', line_color='blue', name='Media Movel')

fig = go.Figure(data=[candlestick])
fig.add_trace(m5)
fig.show()
```
